```python
import fbprophet
import pandas as pd
import itertools
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("SN_y_tot_V2.0.csv", sep = ";", names = ['Year',
                                                            'fraction_date',
                                                            'total_mean_sunspot',
                                                            'mean_std',
                                                            'n_observations_used',
                                                            'definitive/provisional'])
```

```python
data = data.loc[data['Year'] >= 2000].reset_index(drop = True)
```

```python
data['Year'] = data['Year'].apply(int)
data['Year'] = data['Year'].apply(str)
```

```python
data['ds'] = data.Year
data['ds']= pd.to_datetime(data['ds'], format='%Y')
data['y'] = data['total_mean_sunspot']
```

```python
train = data.loc[data['Year'].astype(int) <= 2015]
test = data.loc[data['Year'].astype(int) > 2015]
```

```python
data
```

# Train Model

```python
model = fbprophet.Prophet()
# fit the model
model.fit(data)
```

# 1 year forecast

```python
future = model.make_future_dataframe(periods=1, freq='Y')
future = future.loc[future['ds'] > "2021-01-01"].reset_index(drop = True)
forecast = model.predict(future)
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
model.plot(forecast)
plt.show()
```

# 10 years forecast

```python
future = model.make_future_dataframe(periods=10, freq='Y')
future = future.loc[future['ds'] > "2021-01-01"].reset_index(drop = True)
forecast = model.predict(future)
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
model.plot(forecast)
plt.show()
```

# 20 years forecast

```python
future = model.make_future_dataframe(periods=20, freq='Y')
future = future.loc[future['ds'] > "2021-01-01"].reset_index(drop = True)
forecast = model.predict(future)
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
model.plot(forecast)
plt.show()
```

# Tune Model

```python
param_grid = {  
    'changepoint_prior_scale': [0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.1, 1.0, 10.0],
}

cutoffs = pd.to_datetime(['2005-01-01', '2010-01-01', '2015-01-01'])

all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
rmses = []

# Use cross validation to evaluate all parameters
for params in all_params:
    m = fbprophet.Prophet(**params).fit(data)  # Fit model with given params
    df_cv = fbprophet.diagnostics.cross_validation(m, cutoffs=cutoffs, horizon='365 days', parallel="processes")
    df_p = fbprophet.diagnostics.performance_metrics(df_cv, rolling_window=1)
    rmses.append(df_p['rmse'].values[0])

# Find the best parameters
tuning_results = pd.DataFrame(all_params)
tuning_results['rmse'] = rmses
```

```python
print(tuning_results)
```

# Evaluation

```python
m = fbprophet.Prophet(changepoint_prior_scale = 0.5, seasonality_prior_scale = 10)
m.fit(train)
```

```python
future = m.make_future_dataframe(periods=test.shape[0], freq='Y')
forecast_valid = m.predict(future)
```

```python
m.plot(forecast_valid)
plt.show()
```

```python
ax = forecast_valid.plot(x='ds',y='yhat',legend=True,label='predictions',figsize=(12,8))
test.plot(x='ds', y='y', legend=True,label='True Test Data')
```

```python
test = test.loc[test['ds'] <= "2020-01-01"].reset_index(drop = True)
```

```python
forecast_valid = forecast_valid.loc[forecast_valid['ds'] >= "2016-01-01"].reset_index(drop = True)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error

print('r2 test: {}'.format(r2_score(y_true=test['y'], y_pred=forecast_valid['yhat'])))
print('---'*10)
print('mae test: {}'.format(mean_absolute_error(y_true=test['y'], y_pred=forecast_valid['yhat'])))
print('---'*10)
print('mape test: {}'.format(mean_absolute_percentage_error(y_true=test['y'], y_pred=forecast_valid['yhat'])))
```
