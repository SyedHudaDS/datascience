# Exploratory Data Analysis

This goal of this assignment is to understand the logic and methods of exploratory data analysis(EDA). 
The mode of analysis concerned with discovery, exploration, and empirically detecting phenomena in data. EDA had become the default pre-modeling step for every Machine Learning project engagement. Exploratory Data Analysis(EDA) is a way to investigate datasets and preliminary information, insights, or uncover underlying patterns in the data. Instead of making assumptions, data can be processed in a systematic method to gain insights and make informed decisions.

Sweetviz is an open-source Python library that generates beautiful, high-density visualizations to kickstart EDA (Exploratory Data Analysis) with just two lines of code. Output is a fully self-contained HTML application.

The system is built around quickly visualizing target values and comparing datasets. Its goal is to help quick analysis of target characteristics, training vs testing data, and other such data characterization tasks.

### Reading Data Frame by using Pandas.

When we are reading total data using Pandas, there may be wrong values in data.

```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('telco-customer-churn.csv') # read csv into dataframes file using Pandas
n = 3 # the number of data to display
df.head(n)
```

## clean the data

Data contain wrong values such as null values. Therefore, we must be thought unuseful data.

```python
# columns the contains numeric values :  SeniorCitizen,    tenure,    MonthlyCharges,      TotalCharges
df["SeniorCitizen"] = pd.to_numeric(df["SeniorCitizen"],errors='coerce')
ori_shape = df.shape[0]
df = df.dropna()
new_shape = df.shape[0]
print('=======reduced counts : ', ori_shape - new_shape )
df.info()
```

```python
df["tenure"] = pd.to_numeric(df["tenure"],errors='coerce')
ori_shape = df.shape[0]
df = df.dropna()
new_shape = df.shape[0]
print('=======reduced counts : ', ori_shape - new_shape )
df.info()
```

```python
df["MonthlyCharges"] = pd.to_numeric(df["MonthlyCharges"],errors='coerce')
ori_shape = df.shape[0]
df = df.dropna()
new_shape = df.shape[0]
print('=======reduced counts : ', ori_shape - new_shape )
df.info()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors='coerce')
ori_shape = df.shape[0]
df = df.dropna()
new_shape = df.shape[0]
print('=======reduced counts : ', ori_shape - new_shape )
df.info()
```

## Analyzing the function analyze() of SweetViz And making reports

##### Features of sweetviz

Target analysis, Visualize and compare, Mixed-type associations, Type inference, Summary information

```python
# Reporting by SweetViz
import sweetviz as sz
report = sz.analyze(df)

# Reporting
report.show_html("sweetviz_total_report.html", open_browser=False, layout='widescreen', scale=None) # generating to "sweetvis_total_report.html"
```

```python
from IPython.display import IFrame
IFrame(src='sweetviz_total_report.html', width = 1000, height=1500)
```

```python
from sklearn.model_selection import train_test_split
# Spliting total data into train and test datasets for cross-validation(test dataset size : 30%)
train_set, test_set = train_test_split(df, test_size = 0.3)
```

```python
# train and test datasets Shape
print('---train dataset shape : ', train_set.shape, '\n')
print('---test dataset shape  : ', test_set.shape)
```

##### Comparing  train dataset and test dataset by using the funciton compare() of Sweetviz

```python
# Comparing train and test datasets by using compare function of Sweetviz
compare = sz.compare([train_set, "train dataset"], [test_set, "test dataset"])
#show report
compare.show_html("train_test_report.html", open_browser=False, layout='widescreen', scale=None)
```
