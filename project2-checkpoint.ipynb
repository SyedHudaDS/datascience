This goal of this assignment is to understand the logic and methods of exploratory data analysis(EDA).

# Data Preprocessing

from the sweetviz generated report we noticed that there is no missing values so no need for imputation, the steps we need to follow now is the feature selection and scaling and the encoding of categorical variables so our data can be used for machine learning algorithms

## Checking feature types

```python
# displaying categorical features from our data frame
cat_cols = df.select_dtypes(include=['object']).drop("Churn", axis = 1).columns
cat_cols
```

we notice that TotalCharges is considered to be categorical even though it's actually numerical

```python
#df["TotalCharges"] = df["TotalCharges"].astype(float)
```

when we try to convert it into a numerical value we get this error : \
**ValueError: could not convert string to float: ''** \
which means there are values where TotalCharges is an empty string

```python
# There are eleven values which have an empty field for TotalCharges so we can simply drop those rows
df.loc[df["TotalCharges"] == ' '].shape[0]
```

```python
# Dropping the columns with missing TotalCharges
df = df.drop(df.loc[df["TotalCharges"] == ' '].index).reset_index(drop = True)
```

```python
# Converting TotalCharges to numerical
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

## Encoding Categorical Variables

```python
# numerical columns
num_cols = df.select_dtypes(exclude=['object']).columns
```

```python
# getting the categorical columns after fixing the issue
cat_cols = df.select_dtypes(include=['object']).drop(["Churn", "customerID"], axis = 1).columns
```

```python
cat_cols
```

```python
# Encoding features
for col in cat_cols:
    if df[col].value_counts().count() == 2:
        # Converting yes and no categories to binary
        df[col] = df[col].astype("category").cat.codes
    else:
        # One hot encoding for variables with multiple categories
        df = pd.get_dummies(df, columns=[col])
```

```python
# label encoding the target variable
lb = LabelEncoder()
df["Churn"] = lb.fit_transform(df["Churn"])
```

```python
df.shape
```

## Feature Selection

```python
# feature selection with ANOVA (selecting only the first 30 features from 42)
fs = SelectKBest(score_func=f_classif, k=30)
fs.fit(df.drop(["customerID", "Churn"], axis = 1), df["Churn"])
```

```python
# getting the selected feature names
features = fs.get_feature_names_out()
```

```python
features
```

```python
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(35, 25))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(),
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

correlation_heatmap(df[np.append(features, "Churn")])
```

# Train Test Split

```python
# Spliting total data into train and test datasets for cross-validation(test dataset size : 20%)
train_set, test_set = train_test_split(df, test_size = 0.2)
```

```python
# Scaling Numerical features
mm_scaler = MinMaxScaler()
train_set[['tenure', 'MonthlyCharges', 'TotalCharges']] = mm_scaler.fit_transform(train_set[['tenure', 'MonthlyCharges', 'TotalCharges']])
test_set[['tenure', 'MonthlyCharges', 'TotalCharges']] = mm_scaler.transform(test_set[['tenure', 'MonthlyCharges', 'TotalCharges']])
```

# Experiments

## Model Training and Evaluation w/o SMOTE

```python
X_train, y_train, X_test, y_test = train_set[features], train_set["Churn"], test_set[features], test_set["Churn"]
```

```python
# Defining the classifiers
classifiers = {'Naive Bayes' : GaussianNB(), 
               'RandomForest': RandomForestClassifier(), 
               'LogisticRegression' : LogisticRegression(),
               'XGBClassifier' : XGBClassifier()}
```

```python
# Creating a metrics table
metrics = pd.DataFrame(columns = ["Model", "Accuracy", "Precision", "Recall", "F1 Score"])
# Going through all classifiers training them and getting their metrics and adding them to the table
for model_name in classifiers.keys():
    model = classifiers[model_name]
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    metrics.loc[metrics.shape[0]] = [f"{model_name} w/o SMOTE", 
                                     accuracy_score(y_test, y_pred), 
                                     precision_score(y_test, y_pred),
                                     recall_score(y_test, y_pred),
                                     f1_score(y_test, y_pred)]
```

```python
metrics.sort_values(by = 'Recall', ascending = False)
```

```python
metrics.plot(x="Model",y=["Accuracy", "Precision", "Recall", "F1 Score"], kind="bar",figsize=(9,8), colormap="winter")
```

## Model Training and Evaluation with SMOTE

```python
# Performing SMOTE on the train set to deal with class imbalance
smt = SMOTE()
X_train_sm, y_train_sm = smt.fit_resample(X_train, y_train)
```

```python
for model_name in classifiers.keys():
    model = classifiers[model_name]
    model.fit(X_train_sm, y_train_sm)
    y_pred = model.predict(X_test)
    metrics.loc[metrics.shape[0]] = [f"{model_name} with SMOTE", 
                                     accuracy_score(y_test, y_pred), 
                                     precision_score(y_test, y_pred),
                                     recall_score(y_test, y_pred),
                                     f1_score(y_test, y_pred)]
```

```python
# Sorting the metrics table by F1 Score
metrics.sort_values(by = 'Recall', ascending = False)
```

```python
metrics.plot(x="Model",y=["Accuracy", "Precision", "Recall", "F1 Score"], kind="bar",figsize=(15,12), colormap="winter")
```

## Hyperparameter optimization

```python
def objective(trial):
    # define the models and the parametres spaces for each model    
    classifier_name = trial.suggest_categorical('classifier', ['RandomForestClassifier',
                                                               'XGBClassifier'])
    if classifier_name == 'RandomForestClassifier':
        n_estimators = trial.suggest_int('n_estimators', 50, 1000)
        max_depth = trial.suggest_int('max_depth', 4, 50)
        min_samples_split = trial.suggest_int('min_samples_split', 2, 150)
        min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 60)
        classifier_obj = RandomForestClassifier(n_estimators = n_estimators,
                                                max_depth = max_depth,
                                                min_samples_split = min_samples_split,
                                                min_samples_leaf = min_samples_leaf)
    elif classifier_name == "XGBClassifier":
        params = {'objective' : 'binary:logistic',
              'learning_rate' : trial.suggest_float("learning_rate", 0.01, 0.8),
              'n_estimators' : trial.suggest_int("n_estimators", 100, 1000),
              'max_depth' : trial.suggest_int("max_depth", 2, 600),
              'min_child_weight' : trial.suggest_int("min_child_weight", 1, 20),
              'gamma' : trial.suggest_float("gamma", 0.1, 1),
              'subsample' : trial.suggest_float("subsample", 0.1, 1),
              'colsample_bytree' : trial.suggest_float("colsample_bytree", 0.1, 1),
              'use_label_encoder' : False
             }
        classifier_obj = XGBClassifier(**params)
    
    classifier_obj.fit(X_train_sm, y_train_sm)
    y_pred = classifier_obj.predict(X_test)
    recall = recall_score(y_test, y_pred)
    return recall

# Create a study object and optimize the objective function.
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=150)
```

```python
study.best_params
```

```python
model = RandomForestClassifier(n_estimators = 75, max_depth = 6, min_samples_split = 100, min_samples_leaf = 43)
model.fit(X_train_sm, y_train_sm)
y_pred = model.predict(X_test)
metrics.loc[metrics.shape[0]] = ["Random Forest with SMOTE and hyperopt", 
                                 accuracy_score(y_test, y_pred), 
                                 precision_score(y_test, y_pred),
                                 recall_score(y_test, y_pred),
                                 f1_score(y_test, y_pred)]
```

```python
metrics.sort_values(by = 'Recall', ascending = False)
```
